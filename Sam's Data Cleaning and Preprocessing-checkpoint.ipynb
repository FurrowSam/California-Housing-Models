```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, balanced_accuracy_score, accuracy_score, classification_report, precision_score, recall_score, f1_score
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, StandardScaler, TargetEncoder
%matplotlib inline
from matplotlib import pyplot as plt
from sklearn.datasets import make_regression, make_swiss_roll
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor, ExtraTreesRegressor, AdaBoostRegressor
from sklearn.svm import SVR, SVC
import seaborn as sns
import pickle
```

```python
heart_df_2020 = pd.read_csv('heart_2020_cleaned.csv')
heart_df_2022 = pd.read_csv('heart_2022_cleaned.csv')
```

```python
heart_df_2022.columns
```

```python
heart_df_2020.columns
```

```python
heart_df_2020_yes = heart_df_2020[heart_df_2020['HeartDisease'] == 'Yes']
heart_df_2020_no = heart_df_2020[heart_df_2020['HeartDisease'] == 'No']
```

```python
heart_df_2020_yes.head()
```

```python
heart_df_2020_no.head()
```

```python
min_count = min(len(heart_df_2020_yes), len(heart_df_2020_no))
min_count
```

```python
balanced_heart_disease_yes = heart_df_2020_yes.sample(n=min_count, random_state=42)
balanced_heart_disease_no = heart_df_2020_no.sample(n=min_count, random_state=42)
```

```python
balanced_training_dataset = pd.concat([balanced_heart_disease_yes, balanced_heart_disease_no]).reset_index(drop=True)
```

```python
train_set, test_set = train_test_split(balanced_training_dataset, test_size=0.2, random_state=42)
```

```python
len(train_set), len(test_set)
```

```python
train_set
```

```python
encoded_dataset = pd.get_dummies(train_set, drop_first=True)  
encoded_dataset.head(), encoded_dataset.shape
```

```python
# Define the order for age categories based on the dataset's unique values
age_category_order = ['18-24', '25-29', '30-34', '35-39', '40-44', '45-49',
                      '50-54', '55-59', '60-64', '65-69', '70-74', '75-79', '80 or older']

# Map the age categories to ordinal values
age_category_mapping = {category: index for index, category in enumerate(age_category_order)}
train_set['AgeCategoryOrdinal'] = train_set['AgeCategory'].map(age_category_mapping)

# Drop the original 'AgeCategory' column
train_set_ordinal = train_set.drop('AgeCategory', axis=1)

# Reapply one-hot encoding to the remaining non-numeric columns, excluding 'AgeCategory'
encoded_dataset_with_ordinal_age = pd.get_dummies(train_set_ordinal, drop_first=True)

# Display the first few rows of the dataset to verify the transformation
display(encoded_dataset_with_ordinal_age.head())
display(encoded_dataset_with_ordinal_age.shape)
```

```python
X = encoded_dataset_with_ordinal_age.drop('HeartDisease_Yes', axis=1)
y = encoded_dataset_with_ordinal_age['HeartDisease_Yes']
```

```python
# Apply the one-hot encoding
X_encoded = pd.get_dummies(X, drop_first=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model = SVC(kernel='linear')
model.fit(X_train_scaled, y_train)
```

```python
linear_svc_predictions = model.predict(X_test_scaled)

# Review the predictions
linear_svc_predictions
```

```python
print('Linear SVC Train Accuracy: %.3f' % model.score(X_train_scaled, y_train))
print('Linear SVC Test Accuracy: %.3f' % model.score(X_test_scaled, y_test))
print('Linear SVC Accuracy Score: %.3f' % accuracy_score(y_test, linear_svc_predictions))
```

```python
print("Confusion Matrix:\n", confusion_matrix(y_test, linear_svc_predictions))
```

```python
tn, fp, fn, tp = confusion_matrix(y_test, linear_svc_predictions).ravel()
print(f"False Positive Rate: {fp / (fp + tn)}")
print(f"False Negative Rate: {fn / (fn + tp)}")
```

```python
print(f"True Positive Rate: {tp / (tp + fn)}")
print(f"True Negative Rate: {tn / (tn + fp)}")
```

```python
cm = confusion_matrix(y_test, linear_svc_predictions)

plt.figure(figsize=(10,8))
sns.heatmap(cm, annot=True, fmt="d", cmap="Reds")
plt.title("Confusion Matrix")
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```

```python
print(classification_report(y_test, linear_svc_predictions))
```

```python
print(f'Accuracy: {accuracy_score(y_test, linear_svc_predictions)}')
print(f'Precision: {precision_score(y_test, linear_svc_predictions)}')
print(f'Recall: {recall_score(y_test, linear_svc_predictions)}')
print(f'F1 Score: {f1_score(y_test, linear_svc_predictions)}')
```
